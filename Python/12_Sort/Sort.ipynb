```python
# compare 7 different sort algorithms.

# bubble sort
def bubble_sort(sorted_list):
    list_length = len(sorted_list)
    for i in range(0, list_length-1):
        for j in range(0, list_length-i):
            if(sorted_list[list_length-j-2] > sorted_list[list_length-j-1]):
                sorted_list[list_length-j-2], sorted_list[list_length-j-1] = sorted_list[list_length-j-1], sorted_list[list_length-j-2]
    return sorted_list
                
# selection sort
def selection_sort(sorted_list):
    list_length = len(sorted_list)
    for i in range(0, list_length):
        min_index = i
        for j in range(i, list_length - 1):
            if sorted_list[min_index] > sorted_list[j + 1]:
                min_index = j + 1
        sorted_list[i], sorted_list[min_index] = sorted_list[min_index], sorted_list[i]
    return sorted_list

# insertion sort
def insertion_sort(sorted_list):
    list_length = len(sorted_list)
    for i in range(0, list_length):
        temp = sorted_list[i]
        j = i - 1
        while(j >= 0 and temp < sorted_list[j]):
            sorted_list[j+1] = sorted_list[j]
            j -= 1
        sorted_list[j+1] = temp
    return sorted_list

# shell sort
def shell_sort(sorted_list):
    list_length = len(sorted_list)
    # 初始步长
    gap = list_length // 2
    while gap > 0:
        for i in range(gap, list_length):
            # 每个步长进行插入排序
            temp = sorted_list[i]
            j = i
            # 插入排序
            while(j >= gap and sorted_list[j - gap] > temp):
                sorted_list[j] = sorted_list[j - gap]
                j -= gap
            sorted_list[j] = temp
        # 得到新的步长
        gap = gap // 2
    return sorted_list

# merge sort
def merge(left, right):
    result = []
    while(left and right):
        if(left[0] >= right[0]):
            result.append(right.pop(0))
        else:
            result.append(left.pop(0))
    if left:
        result += left
    if right:
        result += right
    return result
        
def merge_sort(sorted_list):
    list_length = len(sorted_list)
    if list_length <= 1:
        return sorted_list
    mid = list_length // 2
    left = sorted_list[:mid]
    right = sorted_list[mid:]
    left = merge_sort(left)
    right = merge_sort(right)
    sorted_list = merge(left, right)
    return sorted_list
    
# quick sort
def partition(lst, start, end):
    i = start - 1
    pivot_index = end
    pivot = lst[end]
    for j in range(start, end):
        if lst[j] < pivot:
            i = i + 1
            lst[i], lst[j] = lst[j], lst[i]
    lst[i + 1], lst[pivot_index] =  lst[pivot_index], lst[i + 1]
    return i + 1

def sort(lst, start, end):
    if start >= end:
        return
    p = partition(lst, start, end)
    sort(lst, start, p - 1)
    sort(lst, p + 1, end)
    
def quick_sort(sorted_list):
    list_length = len(sorted_list)
    sort(sorted_list, 0, list_length - 1)
    return sorted_list
    
# heap sort

def heap_sort(sorted_list):
    def sift_down(start, end):
        root = start
        while True:
            child = 2 * root + 1
            if child > end:
                break
            if child + 1 <= end and sorted_list[child] < sorted_list[child + 1]:
                child += 1
            if sorted_list[root] < sorted_list[child]:
                sorted_list[root], sorted_list[child] = sorted_list[child], sorted_list[root]
                root = child
            else:
                break
    list_length = len(sorted_list)
    for start in range(list_length - 2):
        sift_down(start, list_length - 1)
    for end in range(list_length - 1, 0, -1):
        sorted_list[0], sorted_list[end] = sorted_list[end], sorted_list[0]
        sift_down(0, end - 1)
    return sorted_list
    
```
